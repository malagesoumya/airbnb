# Data Cleaning and Preparation:


```python
# find null values from dataset
```

```python
data.isnull().sum()          # there is non-null values
```

```python
data['STATE_UT_NAME'] = data['STATE_UT_NAME'].str.upper()
```

```python
# Remove Duplicates:
```

```python
data.duplicated().sum()        # there are non-duplicate values
```

```python
data['MONSOON_TOTAL'] = data['JUN'] +data['JUL'] + data['AUG'] + data['SEP']
```

# Descriptive Analysis:

```python
print('SUMMARY OF DATASET')
data.describe(include = 'all')
```

```python
# VISUALIZATION OF DATASET
```

```python
print(" VISUALIZATION OF THE DATASET")
# Set the style for the plots
sns.set(style="whitegrid")

# Histogram of annual rainfall
plt.figure(figsize=(10, 6))
sns.histplot(data['ANNUAL'], bins=30, kde=True)
plt.title('Annual Rainfall')
plt.xlabel('Annual Rainfall (mm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot of monthly rainfall
plt.figure(figsize=(12, 8))
sns.boxplot(data=data[['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']])
plt.title('Monthly Rainfall')
plt.xlabel('Month')
plt.ylabel('Rainfall (mm)')
plt.show()
```

```python
# Scatter plot of January vs. July rainfall
plt.figure(figsize=(10, 6))
sns.scatterplot(x='JAN', y='JUL', data=data)
plt.title('January vs. July Rainfall')
plt.xlabel('January Rainfall (mm)')
plt.ylabel('July Rainfall (mm)')
plt.show()
```

```python
# Heat map of rainfall distribution
pivot_table = data.pivot_table(values='ANNUAL', index='DISTRICT', columns='STATE_UT_NAME')
plt.figure(figsize=(14, 10))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='.1f')
plt.title('Annual Rainfall Distribution')
plt.xlabel('State/UT')
plt.ylabel('District')
plt.show()
```

```python
# Box plot comparing annual rainfall across states
plt.figure(figsize=(16, 8))
sns.boxplot(x='STATE_UT_NAME', y='ANNUAL', data=data)
plt.xticks(rotation=90)
plt.title('Comparison of Annual Rainfall Across States')
plt.xlabel('State/UT')
plt.ylabel('Annual Rainfall (mm)')
plt.show()
```

```python
# Assuming you have seasonal columns like 'Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec'
plt.figure(figsize=(14, 8))
sns.boxplot(data=data[['Jan-Feb', 'Mar-May', 'Jun-Sep', 'Oct-Dec']])
plt.title('Box Plot of Seasonal Rainfall')
plt.xlabel('Season')
plt.ylabel('Rainfall (mm)')
plt.show()
```

# Trend Analysis:

```python
# average monthly rainfall from datset
```

```python

monthly_avg = data[['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']].mean()

# Plot average monthly rainfall
plt.figure(figsize=(14, 7))
monthly_avg.plot(kind='bar')
plt.title('Average Monthly Rainfall')
plt.xlabel('Month')
plt.ylabel('Average Rainfall (mm)')
plt.xticks(rotation=45)
plt.show()


```

```python
# average seasonal rainfall
```

```python
data['Winter'] = data[['JAN', 'FEB']].mean(axis=1)
data['Summer'] = data[['MAR', 'APR', 'MAY']].mean(axis=1)
data['Monsoon'] = data[['JUN', 'JUL', 'AUG', 'SEP']].mean(axis=1)
data['Post-Monsoon'] = data[['OCT', 'NOV', 'DEC']].mean(axis=1)
seasonal_avg = data[['Winter', 'Summer', 'Monsoon', 'Post-Monsoon']].mean()

plt.figure(figsize=(14, 7))
seasonal_avg.plot(kind='bar')
plt.title('Average Seasonal Rainfall')
plt.xlabel('Season')
plt.ylabel('Average Rainfall (mm)')
plt.xticks(rotation=0)
plt.show()
```

# Regional Analysis:

```python
#  Highest and Lowest Rainfall
```

```python

max_rainfall = data.loc[data['ANNUAL'].idxmax()]
print(f"District with the highest annual rainfall:\n{max_rainfall}")


```

```python
# the district with the lowest annual rainfal
```

```python

min_rainfall = data.loc[data['ANNUAL'].idxmin()]
print(f"District with the lowest annual rainfall:\n{min_rainfall}")
```

```python
# average annual rainfall for each state
```

```python

state_avg_rainfall = data.groupby('STATE_UT_NAME')['ANNUAL'].mean().sort_values(ascending=False)

# Plot average annual rainfall by state
plt.figure(figsize=(14, 7))
state_avg_rainfall.plot(kind='bar')
plt.title('Average Annual Rainfall by State')
plt.xlabel('State/UT')
plt.ylabel('Average Annual Rainfall (mm)')
plt.xticks(rotation=90)
plt.show()

# Box plot of annual rainfall by state
plt.figure(figsize=(16, 8))
sns.boxplot(x='STATE_UT_NAME', y='ANNUAL', data=data)
plt.xticks(rotation=90)
plt.title('Comparison of Annual Rainfall Across States')
plt.xlabel('State/UT')
plt.ylabel('Annual Rainfall (mm)')
plt.show()
```
